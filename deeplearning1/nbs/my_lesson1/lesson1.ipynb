```python
from vgg16 import Vgg16
vgg = Vgg16()
vbatches = vgg.get_batches (cwd + "/data/validation", batch_size=64)
tbatches = vgg.get_batches (cwd + "/data/train", batch_size=64)
vgg.finetune(tbatches)
vgg.fit(tbatches, vbatches)
```

```python
import time
test_batches = vgg.get_batches(cwd + "/data/test", batch_size=100, shuffle=False )
fns = test_batches.filenames
stripped = list(map(lambda f: f.lstrip('unknown/').rstrip('.jpg'), fns))
acc_idxs = []
for j in range(125):
    imgs,labels = next(test_batches)
    preds,idxs,classes = vgg.predict(imgs)
    for idx in idxs:
        acc_idxs.append(idx)
    
final = zip(stripped, acc_idxs)
f = open('final2347.csv', 'w')
for i in final:
    output_string='{},{}.0'.format(i[0],i[1])
    print(output_string)
    print(output_string,file=f)
    
f.flush
time.sleep(2)
f.flush
time.sleep(2)
f.close
time.sleep(2)
```
